# Creating a User Database

```python
import csv
from IPython.display import clear_output

FILE_NAME = "user_db.csv"
fail_message = "Incorrect email and/or password."
no_command = "Command not found."

def getEmail(): return input("Enter email: ").lower()

def getPassword(): return input("Enter password: ")
    
def getPassword2(): return input("Re-enter password: ")

def createDatabase():
    try:
        with open(FILE_NAME, mode="x", newline="") as f:
            writer = csv.writer(f, delimiter=",")
            writer.writerow(["Email", "Password"])      
    except:
        pass

def canLogIn(email, pw):
    if email == "Email":
        return False
    else:
        with open(FILE_NAME, mode="r") as f:
            reader = csv.reader(f, delimiter=",")
            found = False
            for row in reader:
                if email in row and pw in row:
                    found = True
                    break
            return True if found else False

def logIn(email):
    clear_output()
    print("Logging in...\n")
    print("Welcome, {}.".format(email.split("@")[0]).title())
                
def attemptRegister():
    accepted = False
    while not accepted:
        email = getEmail()
        pw = getPassword()
        pw2 = getPassword2()
        if pw != pw2:
            print("Passwords must match.")
        else:
            accepted = True
            if not alreadyRegistered(email):
                register(email, pw)
                print("Registration confirmed!")
            else:
                print("That email is already registered.")
                
def alreadyRegistered(email):
    with open(FILE_NAME, mode="r") as f:
        reader = csv.reader(f, delimiter=",")
        found = False
        for row in reader:
            if email in row:
                found = True
                break
        return True if found else False
    
def register(email, pw):
    with open(FILE_NAME, mode="a", newline="") as f:
        writer = csv.writer(f, delimiter=",")
        writer.writerow([email, pw])
        
def main():
    done = False
    logged_in = False
    
    createDatabase() 
    
    while not done:
        if logged_in:
            ans = input("Would you like to log out (y/n)? ").lower()
            if ans == "y":
                print("Logging out.")
                logged_in = False
            elif ans == "n":
                print("There's really nothing to do here...")
            else:
                print(no_command)
        else:
            ans = input("l to log in, r to register, q to quit: ").lower()
            if ans == "l":
                email, pw = getEmail(), getPassword()
                if canLogIn(email, pw):
                    logIn(email)
                    logged_in = True
                else:
                    print(fail_message)                    
            elif ans == "r":
                attemptRegister()
            elif ans == "q":
                print("Goodbye.")
                done = True
            else:
                print(no_command)
                           
main()
```
